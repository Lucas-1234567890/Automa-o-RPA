## Desafio

- Vamos automatizar a extração de informações de um sistema e envio de um relatório por e-mail
- No nosso caso, para todo mundo conseguir fazer o mesmo programa, o nosso "sistema" vai ser o Gmail, mas o mesmo processo pode ser feito com qualquer programa do seu computador e qualquer sistema
    - Passo 1: Entrar no sistema (entrar no Gmail)
    - Passo 2: Entrar em uma aba específica do sistema onde tem o nosso relatório (Aba Contatos)
    - Passo 3: Exportar o Relatório (Exportar Contatos)
    - Passo 4: Pegar o relatório exportado, tratar e pegar as informações que queremos
    - Passo 5: Preencher/Atualizar informações do sistema (No nosso caso, criar um e-mail e enviar)

```python
import pyautogui
import time
import pandas as pd
import pyperclip

# pyautogui.write() -> escreve
# pyautogui.click -> clica
# pyautogui.locateOnScreen -> identifica uma imagem na sua tela
# pyautogui.hotkey -> usa atalhos do teclado (combinação de teclas)
# pyautogui.press -> aperta um botão do teclado
# print(pyautogui.KEYBOARD_KEYS)

pyautogui.alert('O código vai começar. Não mexa em NADA enquanto o código tiver rodando. Quando finalizar, eu te aviso')
time.sleep(1)

pyautogui.PAUSE = 0.4
# apertar o windows do teclado
pyautogui.press('win')
# digitar chrome
pyautogui.write("chrome")
# apertar enter
pyautogui.press('enter')
time.sleep(2)
# entrar no Gmail
pyautogui.hotkey('win', 'up')

pyautogui.write('gmail')
pyautogui.press('enter')
time.sleep(3)
#esperar carregar o google
while not pyautogui.locateOnScreen('Captura de tela 2025-05-22 120453.png'):
    time.sleep(1)

# localizar a imagem -> vai te dar 4 informações: posicao x, posicao y, largura e altura
x, y, largura, altura = pyautogui.locateOnScreen('Captura de tela 2025-05-22 120453.png')
# clicar no meio da imagem
pyautogui.click(x + largura/4, y + altura/2)
time.sleep(3)

# entrar em contatos
x, y, largura, altura = pyautogui.locateOnScreen('ponto.png')
pyautogui.click(x + largura/1.5, y + altura/2)

pyautogui.moveRel(0, 100, duration=0.5)

pyautogui.scroll(-1000)

try:
    local = pyautogui.locateOnScreen('contatos.png', confidence=0.8)
    if local:
        x, y, largura, altura = local
        pyautogui.click(x + largura/2, y + altura/2)
    else:
        print("Imagem não encontrada")
except pyautogui.ImageNotFoundException:
    print("Imagem não encontrada na tela")

time.sleep(2)

try:
    exportar = pyautogui.locateOnScreen('exportar.png', confidence=0.8)
    if exportar:
        x, y, largura, altura = exportar
        pyautogui.click(x + largura/2, y + altura/2)
    else:
        print('img nao encontrada')
except pyautogui.ImageNotFoundException:
    print('achei nao')
time.sleep(2)
try:
    exportar = pyautogui.locateOnScreen('confirmar_exportarD.png', confidence=0.8)
    if exportar:
        x, y, largura, altura = exportar
        pyautogui.click(x + largura/1.5, y + altura/2)
    else:
        print('img nao encontrada')
except pyautogui.ImageNotFoundException:
    print('achei nao')


# Carregar o arquivo
df = pd.read_csv(r"C:\Users\Lucas\Downloads\contacts.csv")

# Filtrar contatos com email preenchido
df_filtrado = df[df['E-mail 1 - Value'].notna() & (df['E-mail 1 - Value'] != '')].iloc[:, :19]

# Mostrar
df_filtrado

pyautogui.hotkey('ctrl', 'pgup')

for email in df_filtrado['E-mail 1 - Value']:
    try:
        escrever = pyautogui.locateOnScreen('escrever.png', confidence=0.8)
        if escrever:
            x, y, largura, altura = escrever
            pyautogui.click(x + largura/4, y + altura/1.3)
        else:
            print('img nao encontrada')
    except pyautogui.ImageNotFoundException:
        print('achei nao')
    pyautogui.write(email)
    pyautogui.press('enter')
    pyautogui.press('tab')
    pyautogui.write('Olá, este é um email do pyautogui')
    pyautogui.press('tab')
    texto = """"
    Oi, amor. Espero que eu só tenha o seu contato com email na minha lista, se nao isso vai para um monte de gente
    
    """
    pyperclip.copy(texto)
    pyautogui.hotkey('ctrl', 'v')
    pyautogui.hotkey('ctrl', 'enter')

pyautogui.alert('O codigo terminou, pode pegar o seu pc de volta')
```
